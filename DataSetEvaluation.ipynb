```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Users\Medha\Documents\MINE\learning\Diet_Rec_System\diet_recommendations_dataset.csv")

# Display the first few rows to understand the structure
#print(df.head())
print(df.columns)

# Get the number of rows and columns
#print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
# Check for missing values in each column
print('\n',df.isnull().sum())

# Summary statistics for numerical columns
print(f"Summary statistics:\n{df.describe()}")
```

```python
df['Disease_Type'].fillna('None', inplace=True)
df['Dietary_Restrictions'].fillna('None', inplace=True)
df['Allergies'].fillna('None', inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# Age range
age_min = df['Age'].min()
age_max = df['Age'].max()

# BMI range
bmi_min = df['BMI'].min()
bmi_max = df['BMI'].max()

print(f"Age Range: {age_min} - {age_max}")
print(f"BMI Range: {bmi_min:.2f} - {bmi_max:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
sns.boxplot(data=df, x='Disease_Type', y='BMI', hue='Gender')
plt.title("BMI vs Disease Type colored by Severity")
plt.xticks(rotation=45)
plt.legend(title='Severity', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.stripplot(data=df, x='Disease_Type', y='BMI', hue='Severity', dodge=True, jitter=0.25, alpha=0.7)
plt.title("Individual BMI values by Disease Type with Severity")
plt.xticks(rotation=45)
plt.legend(title='Severity', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
